```python
import numpy as np
import matplotlib.pyplot as plt
```

**(1)** Сперва зададим функцию и ядра свёртки:

```python
def impulse(A, x):
    return (A if (x % 2 >= 1) else -A)
```

```python
def kernel1(A, x):
    return A * (np.exp(-x**2))

def kernel2(A, x):
    # return np.tanh(x % 1/2)
    return A * (x - np.tanh(2*A))
```

Теперь выполним свёртку через `np.convolve`:

```python
print(np.convolve((1,2,3), (4,5,6), 'same'))
print(np.convolve((1,2,3), (4,5,6), 'full'))
```

```python
for krnl in [kernel1, kernel2]:
    time = np.arange(0, 10, 0.1)

    values_fun = [impulse(5, a) for a in time]
    values_krnl = [krnl(5, a) for a in time]

    values_conv = np.convolve(
        values_fun,
        values_krnl,
        'same' # Возвращаем список такой же размерности (обрезаем неполные операции)
    )
    print(values_conv.shape)
    print(values_conv)


    plt.figure(figsize=(8, 12))

    plt.subplot(3, 1, 1)
    plt.plot(time, values_fun)

    plt.subplot(3, 1, 2)
    plt.plot(time, values_krnl)

    plt.subplot(3, 1, 3)
    plt.plot(time, values_conv)

    plt.show()
```
